```python
import numpy as np
import matplotlib.pyplot as plt

# Генерация двух векторов
vector1 = np.array([1, 2])
vector2 = np.array([3, 1])

# Расчет косинусного расстояния
cosine_distance = np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))

# Расчет угла в радианах
angle_rad = np.arccos(cosine_distance)

# Перевод угла из радиан в градусы
angle_deg = np.degrees(angle_rad)

# Визуализация векторов и угла между ними
fig, ax = plt.subplots()

ax.quiver(0, 0, vector1[0], vector1[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
ax.quiver(0, 0, vector2[0], vector2[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')

ax.set_xlim(0, 4)
ax.set_ylim(0, 4)

# Подписи к векторам
ax.text(vector1[0] + 0.1, vector1[1] + 0.1, 'Vector 1', color='r', fontsize=10)
ax.text(vector2[0] + 0.1, vector2[1] + 0.1, 'Vector 2', color='b', fontsize=10)

# Рисование угла между векторами
angle_str = f'Angle: {angle_deg:.2f} degrees\nCosine Distance: {cosine_distance:.2f}'
ax.text(0.5, 3.5, angle_str, fontsize=12)

plt.legend()
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Пример данных
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 3, 4, 5, 6])

# Рассчитываем коэффициент корреляции Пирсона
correlation_coefficient = np.corrcoef(X, Y)[0, 1]

# Строим точечный график
plt.scatter(X, Y, label=f'Корреляция: {correlation_coefficient:.2f}', color='blue')

# Добавляем линию тренда (линию наилучшего соответствия)
z = np.polyfit(X, Y, 1)
p = np.poly1d(z)
plt.plot(X, p(X), color='red', linestyle='--')

# Добавляем подписи и легенду
plt.xlabel('Переменная X')
plt.ylabel('Переменная Y')
plt.title('Иллюстрация корреляции Пирсона')
plt.legend()

# Отображаем график
plt.show()
```
